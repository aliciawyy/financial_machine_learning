```python
import pandas as pd
import numpy as np

import fml
%matplotlib inline
```

We compute here the number of labels concurrent at t.

```python
df_price = fml.DataHandler().get_time_series_data("SPY")
price = fml.Price(df_price["Adj Close"], freq="B")
```

```python
cross_time = price.bounds_cross_time(window=10)
```

```python
cross_time.index = range(len(cross_time))
cross_time.head()
```

```python
cross_time_int = cross_time.min(axis=1).to_frame("cross")
cross_time_int["cnt"] = 0
cross_time_int.head(10)
```

```python
for i in cross_time_int.index[:10]:
    cross_time_int.loc[i + 1: i + cross_time_int.loc[i, "cross"], "cnt"] += 1
cross_time_int.head(10)
```
